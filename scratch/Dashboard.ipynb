```python
import datetime
import numpy as np
import pandas as pd

import plotly.graph_objects as go
from ipywidgets import widgets
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/ThuwarakeshM/geting-started-with-plottly-dash/main/life_expectancy.csv"
)
df = df.drop(df.columns[[0]], axis=1)
```

```python
#to build all unique widgets will be using for our data
age = widgets.IntSlider(
    value=1.0,
    min=1.0,
    max=100.0,
    step=10.0,
    description='Age:',
    continuous_update=False
)

use_year = widgets.Checkbox(
    description='Year: ',
    value=True,
)

container = widgets.HBox(children=[use_year, age])

textbox = widgets.Dropdown(
    description='Continent:   ',
    value='Americas',
    options=df['continent'].unique().tolist()
)

origin = widgets.Dropdown(
    options=list(df['Status'].unique()),
    value='Developing',
    description='Developing Country:',
)


# Assign an empty figure widget with two traces
trace1 = go.Scatter(x=df['GDP'], opacity=0.75, name='GDP')
trace2 = go.Histogram(x=df['Population'], opacity=0.75, name='Population')
g = go.FigureWidget(data=[trace1, trace2],
                    layout=go.Layout(
                        title=dict(
                            text='GDP DataBase'
                        ),
                        barmode='overlay'
                    ))
```
